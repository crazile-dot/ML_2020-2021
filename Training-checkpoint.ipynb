```python
#from keras.models import Sequential
#from keras.layers.core import Dense, Dropout, Activation, Flatten
#from keras.layers.convolutional import Convolution2D, MaxPooling2D
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
```

```python
x_data = pd.read_csv('train_gesture_x.csv')
y_data = pd.read_csv('train_gesture_y.csv')
z_data = pd.read_csv('train_gesture_z.csv')
label = pd.read_csv('train_label.csv')
```

```python
ncols = len(z_data.columns)
x_col_names = []*ncols
y_col_names = []*ncols
z_col_names = []*ncols
label_col_names = []*1

x_first_row = []*ncols
y_first_row = []*ncols
z_first_row = []*ncols
label_first_row = []*1

for i in range(ncols):
    x_col_names.append('x_t' + str(i))

x_first_row = x_data.columns
x_data.loc[-1] = x_first_row
x_data.index = x_data.index + 1 
x_data = x_data.sort_index()
x_data.columns = x_col_names

for i in range(ncols):
    y_col_names.append('y_t' + str(i))
    
y_first_row = y_data.columns
y_data.loc[-1] = y_first_row
y_data.index = y_data.index + 1 
y_data = y_data.sort_index()
y_data.columns = y_col_names

for i in range(ncols):
    z_col_names.append('z_t' + str(i))
    
z_first_row = z_data.columns
z_data.loc[-1] = z_first_row
z_data.index = z_data.index + 1 
z_data = z_data.sort_index()
z_data.columns = z_col_names

label_col_names.append('label')
label_first_row = label.columns
label.loc[-1] = label_first_row
label.index = label.index + 1 
label = label.sort_index()
label.columns = label_col_names

df = x_data.join(y_data)
df = df.join(z_data)
df = df.join(label)
```
